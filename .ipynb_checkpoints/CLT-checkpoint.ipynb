```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(0,100) for i in range(10000)]
pop_table = pd.DataFrame({"Pop_data":pop_data})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop_data",data=pop_table) #Non distributed data
plt.show()
```

```python
sample_mean = []
for noof_sampple in range(60): #number of sample = 50
    sample_data = []
    for data in range(500): #500 data in each sample
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
Sample_M = pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Sample_Mean",data=Sample_M) #see the normal distribution
plt.show()
```

```python
np.mean(pop_data) #Original Population Mean
```

```python
np.mean(sample_mean) #Mean of ALl Sample's Mean
```
