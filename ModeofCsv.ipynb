```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("employee.csv",encoding='ISO-8859-1')
```

```python
mo = dataset["Age"].mode()[0] #using pandas
```

```python
dataset["Age"].value_counts() #count no of frequency
```

```python
sns.histplot(x="Age",data=dataset,bins=[i for i in range(0,71,10)])
plt.plot([mo for i in range(0,300)], [i for i in range(0,300)],c="red",label="Mode")
plt.legend()
plt.show()
```
