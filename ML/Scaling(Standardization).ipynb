```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("carsales.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Odometer (KM)"].fillna(dataset["Odometer (KM)"].mean(),inplace=True)
dataset["Price"].fillna(dataset["Price"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Odometer (KM)"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["Odometer (KM)"]])
```

```python
dataset["Odometer (KM)_SS"] = ss.transform(dataset[["Odometer (KM)"]])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1,2,1) # row,column and plotno
plt.title("Before")
sns.distplot(dataset["Odometer (KM)"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Odometer (KM)_SS"])

plt.subplots_adjust(wspace=0.5)

plt.show()
```
