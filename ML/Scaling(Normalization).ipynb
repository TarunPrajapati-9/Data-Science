```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("carsales.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset["Odometer (KM)"].fillna(dataset["Odometer (KM)"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Odometer (KM)"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["Odometer (KM)"]])
```

```python
dataset["Odometer (KM)_Min_Max"]=ms.transform(dataset[["Odometer (KM)"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Odometer (KM)"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Odometer (KM)_Min_Max"])

plt.show()
```
