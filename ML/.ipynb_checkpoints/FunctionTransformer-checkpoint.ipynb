```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("carsales.csv")
dataset.head(5)
```

```python
sns.distplot(dataset["Odometer (KM)"])
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
# ft = FunctionTransformer(func= lambda x: x**2)
ft = FunctionTransformer(func= np.log1p)
```

```python
ft.fit(dataset["Odometer (KM)"])
```

```python
dataset["Odometer (KM)_FTRa"] = ft.transform(dataset["Odometer (KM)"])
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Odometer (KM)"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Odometer (KM)_FTRa"])
plt.show()
```
