```python
import pandas as pd
```

```python
df =  pd.DataFrame({"size": ["s","m","xl","m","l","m","s","l","xl","m","l","m","s"]})
df
```

```python
ord_data = [["s","m","l","xl"]] #order array in which we are going to encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["size"]])
```

```python
df["size_encoded"] = oe.transform(df[["size"]])
df
```

```python
ord_data_map = {"s":0,"m":1,"l":2,"xl":3} #create dictionary to assign any numeric value like 200,400etc.
```

```python
df["size_encoding_map"] = df["size"].map(ord_data_map)
df
```
