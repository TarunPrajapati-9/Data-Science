```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("salary.csv")
dataset.head(3)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset["Level "],dataset["Salary"])
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
x = dataset[["Level "]]
y = dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
pre = lr.predict(x)
```

```python
plt.scatter(dataset["Level "],dataset["Salary"])
plt.plot(dataset["Level "],pre,color="red")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(["Original","Predict Line"])
plt.show()
```

```python
test = pf.transform([[9]])
```

```python
lr.predict(test)
```

```python
data = pd.read_csv("Ice_cream selling data.csv")
data.head(3)
```

```python
data.corr()
```

```python
plt.scatter(data["Temperature (°C)"],data["Ice Cream Sales (units)"])
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```

```python
a = data[["Temperature (°C)"]]
b = data["Ice Cream Sales (units)"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf1 = PolynomialFeatures(degree=2)
pf1.fit(a)
a = pf1.transform(a)
```

```python
from sklearn.model_selection import train_test_split
```

```python
a_train, a_test, b_train , b_test = train_test_split(a,b,test_size=0.2,random_state=42)
```

```python
lr1 = LinearRegression()
lr1.fit(a_train,b_train)
```

```python
lr1.score(a_test,b_test)*100
```

```python
pre = lr1.predict(a)
```

```python
plt.scatter(data["Temperature (°C)"],data["Ice Cream Sales (units)"])
plt.plot(data["Temperature (°C)"],pre,color="red")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.legend(["Original","Predict Line"])
plt.show()
```
